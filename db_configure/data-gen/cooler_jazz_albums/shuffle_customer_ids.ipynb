```python
# extracting ids
customer_ids = customers["customer_id"].tolist()
print(customer_ids)
```

```python
# randomly shuffling these customer ids into orders and reviews
import random 
def random_id(list_ids):
    """ Returns a random id from the list """
    return random.choice(list_ids)

shuffled_orders = orders.copy()
shuffled_reviews = reviews.copy()
shuffled_orders["customer_id"] = orders["customer_id"].apply(lambda x: random_id(customer_ids))
shuffled_reviews["customer_id"] = reviews["customer_id"].apply(lambda x: random_id(customer_ids))
```

# Safe Customer ID Shuffle Migration

This section migrates the shuffled customer IDs to Supabase safely.

**What this does:**
- Updates only the `customer_id` column in orders and reviews
- Keeps the same order_id and review_id (no cascade deletions)
- Preserves all other data and relationships

**Steps:**
1. Preview changes
2. Connect to database
3. Run update
4. Verify results

## Step 1: Preview Changes

Let's see what will be updated before making any changes.

```python
# Compare old vs new customer IDs for orders
orders_comparison = pd.merge(
    orders[['order_id', 'customer_id']],
    shuffled_orders[['order_id', 'customer_id']],
    on='order_id',
    suffixes=('_old', '_new')
)

# Find changes in orders
orders_changes = orders_comparison[orders_comparison['customer_id_old'] != orders_comparison['customer_id_new']]

print(f"üìä Orders Change Summary:")
print(f"   Total orders: {len(shuffled_orders)}")
print(f"   Customer ID changes: {len(orders_changes)}")
print()

# Show some examples
if len(orders_changes) > 0:
    print("üîÄ Sample Order Customer ID Changes (first 10):")
    for idx, row in orders_changes.head(10).iterrows():
        print(f"   Order {row['order_id'][:8]}...")
        print(f"      Old customer: {row['customer_id_old'][:8]}...")
        print(f"      New customer: {row['customer_id_new'][:8]}...")
        print()
```

```python
# Compare old vs new customer IDs for reviews
reviews_comparison = pd.merge(
    reviews[['review_id', 'customer_id']],
    shuffled_reviews[['review_id', 'customer_id']],
    on='review_id',
    suffixes=('_old', '_new')
)

# Find changes in reviews
reviews_changes = reviews_comparison[reviews_comparison['customer_id_old'] != reviews_comparison['customer_id_new']]

print(f"üìä Reviews Change Summary:")
print(f"   Total reviews: {len(shuffled_reviews)}")
print(f"   Customer ID changes: {len(reviews_changes)}")
print()

# Show some examples
if len(reviews_changes) > 0:
    print("üîÄ Sample Review Customer ID Changes (first 10):")
    for idx, row in reviews_changes.head(10).iterrows():
        print(f"   Review {row['review_id'][:8]}...")
        print(f"      Old customer: {row['customer_id_old'][:8]}...")
        print(f"      New customer: {row['customer_id_new'][:8]}...")
        print()
```

## Step 2: Connect to Database

```python
# Initialize database connection
db = DatabaseConnector()
db.connect()

print("‚úÖ Connected to database")
```

## Step 3: Update Orders Table

‚ö†Ô∏è **This will modify your database!** Make sure you've reviewed the preview above.

```python
# Update orders in batches - ONLY customer_id
BATCH_SIZE = 50
total_updated = 0
total_failed = 0

print(f"üîÑ Starting update of {len(shuffled_orders)} orders...")
print(f"‚ö†Ô∏è  Updating ONLY 'customer_id' column")
print(f"‚úÖ Preserving: order_id, order_number, total, shipping_address, order_date, timestamps\n")

for i in range(0, len(shuffled_orders), BATCH_SIZE):
    batch = shuffled_orders.iloc[i:i+BATCH_SIZE]
    print(f"üì¶ Processing batch {i//BATCH_SIZE + 1} ({len(batch)} records)...")
    
    for idx, row in batch.iterrows():
        try:
            # Update ONLY customer_id - nothing else
            db.client.table('orders').update({
                'customer_id': row['customer_id']
            }).eq('order_id', row['order_id']).execute()
            
            total_updated += 1
            
            # Progress indicator
            if total_updated % 10 == 0:
                print(f"  ‚úì Updated {total_updated}/{len(shuffled_orders)}...")
        
        except Exception as e:
            total_failed += 1
            print(f"  ‚ùå Failed to update order {row['order_id']}: {e}")

print(f"\n{'='*60}")
print(f"‚úÖ Orders Update Complete!")
print(f"{'='*60}")
print(f"‚úì Successfully updated: {total_updated}")
if total_failed > 0:
    print(f"‚ùå Failed updates: {total_failed}")
print(f"{'='*60}")
```

## Step 4: Update Reviews Table

‚ö†Ô∏è **This will modify your database!** Make sure you've reviewed the preview above.

```python
# Update reviews in batches - ONLY customer_id
BATCH_SIZE = 50
total_updated = 0
total_failed = 0

print(f"üîÑ Starting update of {len(shuffled_reviews)} reviews...")
print(f"‚ö†Ô∏è  Updating ONLY 'customer_id' column")
print(f"‚úÖ Preserving: review_id, album_id, rating, review_text, created_at\n")

for i in range(0, len(shuffled_reviews), BATCH_SIZE):
    batch = shuffled_reviews.iloc[i:i+BATCH_SIZE]
    print(f"üì¶ Processing batch {i//BATCH_SIZE + 1} ({len(batch)} records)...")
    
    for idx, row in batch.iterrows():
        try:
            # Update ONLY customer_id - nothing else
            db.client.table('reviews').update({
                'customer_id': row['customer_id']
            }).eq('review_id', row['review_id']).execute()
            
            total_updated += 1
            
            # Progress indicator
            if total_updated % 10 == 0:
                print(f"  ‚úì Updated {total_updated}/{len(shuffled_reviews)}...")
        
        except Exception as e:
            total_failed += 1
            print(f"  ‚ùå Failed to update review {row['review_id']}: {e}")

print(f"\n{'='*60}")
print(f"‚úÖ Reviews Update Complete!")
print(f"{'='*60}")
print(f"‚úì Successfully updated: {total_updated}")
if total_failed > 0:
    print(f"‚ùå Failed updates: {total_failed}")
print(f"{'='*60}")
```

## Step 5: Verify the Updates

Let's check random samples to make sure the updates worked.

```python
# Verify orders by checking random samples
sample_size = 10
sample = shuffled_orders.sample(min(sample_size, len(shuffled_orders)))

print(f"üîç Verifying {len(sample)} random orders:\n")
all_match = True

for idx, row in sample.iterrows():
    # Fetch from database
    result = db.client.table('orders').select(
        'order_id, customer_id'
    ).eq('order_id', row['order_id']).execute()
    
    if result.data:
        db_record = result.data[0]
        matches = db_record['customer_id'] == row['customer_id']
        
        status = "‚úÖ" if matches else "‚ùå"
        print(f"{status} Order ID: {row['order_id'][:8]}...")
        print(f"   CSV:  customer_id = {row['customer_id'][:8]}...")
        print(f"   DB:   customer_id = {db_record['customer_id'][:8]}...")
        print()
        
        if not matches:
            all_match = False
    else:
        print(f"‚ùå Order {row['order_id']} not found in database!")
        all_match = False

if all_match:
    print("\n‚úÖ All sampled orders match! Update successful.")
else:
    print("\n‚ö†Ô∏è  Some orders don't match. Please investigate.")
```

```python
# Verify reviews by checking random samples
sample_size = 10
sample = shuffled_reviews.sample(min(sample_size, len(shuffled_reviews)))

print(f"üîç Verifying {len(sample)} random reviews:\n")
all_match = True

for idx, row in sample.iterrows():
    # Fetch from database
    result = db.client.table('reviews').select(
        'review_id, customer_id'
    ).eq('review_id', row['review_id']).execute()
    
    if result.data:
        db_record = result.data[0]
        matches = db_record['customer_id'] == row['customer_id']
        
        status = "‚úÖ" if matches else "‚ùå"
        print(f"{status} Review ID: {row['review_id'][:8]}...")
        print(f"   CSV:  customer_id = {row['customer_id'][:8]}...")
        print(f"   DB:   customer_id = {db_record['customer_id'][:8]}...")
        print()
        
        if not matches:
            all_match = False
    else:
        print(f"‚ùå Review {row['review_id']} not found in database!")
        all_match = False

if all_match:
    print("\n‚úÖ All sampled reviews match! Update successful.")
else:
    print("\n‚ö†Ô∏è  Some reviews don't match. Please investigate.")
```
