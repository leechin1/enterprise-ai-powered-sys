# Changing the album's names
Unfortunately, Gemini's capcity has not been successful in generating a variety of interesting albums. Upon realising this, the data had already been put in the db so I will proceed to tweek this while keeping the respective constraints.


```python
import pandas as pd
```

```python
# Data read
df = pd.read_csv("/Users/leechin/Documents/enterprise-ai-powered-sys/data-gen/cooler_jazz_albums/albums_rows.csv", index_col ="album_id")

print(df.shape)
df.head()
# I only need to change the title and the artist names
```

```python
df_cooler = pd.read_csv("/Users/leechin/Documents/enterprise-ai-powered-sys/data-gen/cooler_jazz_albums/jazz_records.csv")
print(df_cooler.shape)
df_cooler.head(30)
```

```python
df_new = df.copy()
for i, album_id in enumerate(df_new.index):
    df_new.at[album_id, 'title'] = df_cooler.at[i, 'title']
    df_new.at[album_id, 'artist'] = df_cooler.at[i, 'artist']
df_new.head()
```

```python
df_new.to_csv("/Users/leechin/Documents/enterprise-ai-powered-sys/data-gen/cooler_jazz_albums/new.csv", index=True)
```
